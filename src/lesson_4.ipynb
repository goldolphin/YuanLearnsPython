# 平面直角坐标、goto() 函数

1. 直线行走、平移

```python
from ColabTurtle.Turtle import *
initializeTurtle(10)

def reset(x, y):
    penup()
    goto(x, y)
    pendown()

reset(0, 0)
for i in range(50):
    goto(i*10, i*5)

reset(0, 100)
for i in range(50):
    goto(i*10, i*5 + 100)

reset(100, 0)
for i in range(50):
    goto(i*10 + 100, i*5)
```

2. 抛物线

```python
initializeTurtle(10)

reset(0, 0)
for i in range(20):
    goto(i*10, i*i)

reset(0, 0)
for i in range(20):
    goto(i*20, i*i)

reset(0, 0)
for i in range(20):
    goto(i*40, i*i)

reset(0, 0)
for i in range(20):
    goto(i*80, i*i)

reset(0, 0)
for i in range(20):
    goto(i*400, i*i)
```

3. 波浪线

```python
from math import sin
initializeTurtle(10)

reset(0, 100)
for i in range(80):
    goto(i*10, sin(i*0.2)*50 + 100)
```
